```python
def task1():
    conn = psycopg2.connect(**DB_CONFIG)
    cur = conn.cursor()
    
    # Удаляем старую таблицу если существует
    cur.execute("DROP TABLE IF EXISTS logs_corrected;")
    
    # Создаем таблицу с исправленными данными
    cur.execute('''
    CREATE TABLE logs_corrected AS 
    SELECT 
        courseid, userid, num_week, s_all,
        REPLACE(s_all_avg, ',', '.') as s_all_avg,
        s_course_viewed,
        REPLACE(s_course_viewed_avg, ',', '.') as s_course_viewed_avg,
        s_q_attempt_viewed,
        REPLACE(s_q_attempt_viewed_avg, ',', '.') as s_q_attempt_viewed_avg,
        s_a_course_module_viewed,
        REPLACE(s_a_course_module_viewed_avg, ',', '.') as s_a_course_module_viewed_avg,
        s_a_submission_status_viewed,
        REPLACE(s_a_submission_status_viewed_avg, ',', '.') as s_a_submission_status_viewed_avg,
        NameR_Level, Name_vAtt, Depart, Name_OsnO, Name_FormOPril, LevelEd, Num_Sem, Kurs, Date_vAtt
    FROM logs;
    ''')
    conn.commit()
    
    # Выводим первые 10 записей
    cur.execute("SELECT * FROM logs_corrected LIMIT 10;")
    records = cur.fetchall()
    
    print("Первые 10 записей после замены запятых на точки:")
    for record in records:
        print(record)
    
    cur.close()
    conn.close()

task1()
```

```python
# Задание 3
def task3():
    conn = psycopg2.connect(**DB_CONFIG)
    cur = conn.cursor()
    cur.execute('''
        SELECT depart, COUNT(DISTINCT courseid) as course_count 
        FROM logs_corrected 
        GROUP BY depart 
        ORDER BY course_count DESC 
        LIMIT 1;
    ''')
    result = cur.fetchone()
    print(f"Задание 3: Кафедра {result[0]} - {result[1]} курсов (максимум)")
    cur.close()
    conn.close()

task3()
```

```python
def task5():
    conn = psycopg2.connect(**DB_CONFIG)
    cur = conn.cursor()
    cur.execute('''
        SELECT d.name, COUNT(DISTINCT l.courseid) as course_count 
        FROM logs_corrected l
        JOIN departments d ON l.depart = d.id
        GROUP BY d.name 
        ORDER BY course_count DESC 
        LIMIT 1;
    ''')
    result = cur.fetchone()
    print(f"Задание 5: Кафедра {result[0]} - {result[1]} курсов")
    cur.close()
    conn.close()

task5()
```

```python
# Задание 6
def task6():
    conn = psycopg2.connect(**DB_CONFIG)
    cur = conn.cursor()
    
    cur.execute('''
        SELECT COUNT(*) FROM (
            SELECT courseid FROM logs_corrected 
            GROUP BY courseid 
            HAVING COUNT(DISTINCT depart) > 1
        ) as multi_courses;
    ''')
    count = cur.fetchone()[0]
    print(f"Задание 6: Курсов с несколькими кафедрами - {count}")
    
    cur.execute('''
        SELECT l.courseid, STRING_AGG(DISTINCT d.name, ', ') 
        FROM logs_corrected l
        JOIN departments d ON l.depart = d.id
        GROUP BY l.courseid 
        HAVING COUNT(DISTINCT d.name) > 1 
        LIMIT 3;
    ''')
    print("Примеры курсов с несколькими кафедрами:")
    for course in cur.fetchall():
        print(f"  Курс {course[0]}: {course[1]}")
    
    cur.close()
    conn.close()

task6()
```

```python
# Задание 7
def task7():
    conn = psycopg2.connect(**DB_CONFIG)
    cur = conn.cursor()
    cur.execute('''
        SELECT namer_level, COUNT(DISTINCT userid) 
        FROM logs_corrected 
        WHERE namer_level IS NOT NULL 
        GROUP BY namer_level 
        ORDER BY namer_level;
    ''')
    print("Задание 7: Распределение оценок")
    for grade, count in cur.fetchall():
        print(f"  Оценка {grade}: {count} студентов")
    cur.close()
    conn.close()

task7()
```

```python
# Задание 8
def task8():
    conn = psycopg2.connect(**DB_CONFIG)
    cur = conn.cursor()
    cur.execute('''
        SELECT userid, SUM(s_all) 
        FROM logs_corrected 
        GROUP BY userid 
        ORDER BY SUM(s_all) DESC 
        LIMIT 1;
    ''')
    result = cur.fetchone()
    print(f"Задание 8: Самый активный студент {result[0]} - {result[1]} действий")
    cur.close()
    conn.close()

task8()
```

```python
# Задание 9
def task9():
    conn = psycopg2.connect(**DB_CONFIG)
    cur = conn.cursor()
    cur.execute('''
        SELECT num_week, ROUND(AVG(s_all), 2) 
        FROM logs_corrected 
        GROUP BY num_week 
        ORDER BY num_week;
    ''')
    print("Задание 9: Среднее количество событий по неделям")
    for week, avg_events in cur.fetchall():
        print(f"  Неделя {week}: {avg_events}")
    cur.close()
    conn.close()

task9()
```
